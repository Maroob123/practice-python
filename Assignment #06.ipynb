```python
#Q5: Write a code in python in which create a class named it Car which
#have 5 attributes such like (model, color and name etc.) and 3 methods.
#And create 5 object instance from that class. 

class car():
    #five attributes
    def __init__(self,model,color,name,make):
        self.model=model
        self.color=color
        self.name=name
        self.make=make
        self.owner='Syed Muhammad Maroob'
     
    #three methods
    def updatecolor(self,newcolor):
        self.newcolor=newcolor
        print("After update the color of car it becomes",self.newcolor)
        
    def changemodel(self,newmodel):
        self.newmodel=newmodel
        print("After changing the model :", self.newmodel)
        
    def new_owner(self,newowner):
        self.newowner=newowner
        print("After selling Car the newowner is ", self.newowner)

#five objects        
car1=car('2010','black','corola','toyota')
car2=car('2018','white','altis','toyota')
car3=car('2006','blue','platz Automatic','toyota')
car4=car('2014','black','corola XLI','toyota')
car5=car('2013','gray','passo','toyota')

print("For car 01:")
print("Car model is", car1.model)
car1.changemodel("2009")
print("Car color is", car1.color)
car1.updatecolor('red')
print("Car Model is", car1.name)
print("Car is Manufactured by", car1.make)
print("The owner of the car is", car1.owner)
car1.new_owner("Shah Sahab")

print("\n")
print("For car 02:")
print("Car model is", car2.model)
car2.changemodel("2009")
print("Car color is", car2.color)
car2.updatecolor('black')
print("Car Model is", car2.name)
print("Car is Manufactured by", car2.make)
print("The owner of the car is", car2.owner)
car2.new_owner("Shah Sahab")

print("\n")
print("For car 03:")
print("Car model is", car3.model)
car3.changemodel("2009")
print("Car color is", car3.color)
car3.updatecolor('gray')
print("Car Model is", car3.name)
print("Car is Manufactured by", car3.make)
print("The owner of the car is", car3.owner)
car3.new_owner("Shah Sahab")


print("\n")
print("For car 04:")
print("Car model is", car4.model)
car4.changemodel("2009")
print("Car color is", car4.color)
car4.updatecolor('black')
print("Car Model is", car4.name)
print("Car is Manufactured by", car4.make)
print("The owner of the car is", car4.owner)
car4.new_owner("Shah Sahab")

print("\n")
print("For car 05:")
print("Car model is", car5.model)
car5.changemodel("2009")
print("Car color is", car5.color)
car5.updatecolor('red')
print("Car Model is", car5.name)
print("Car is Manufactured by", car5.make)
print("The owner of the car is", car5.owner)
car5.new_owner("Shah Sahab")
```
